## Most Common Artistic Mediums

## Key Insights

- The **Drawings and Prints** department contains the largest share of objects in the Met collection.
- **Egypt** is the most represented country of origin in the dataset.
- Printmaking techniques such as **etching, engraving, and lithography** appear frequently, reflecting the prominence of the museum's print collections.
- Artwork production increases dramatically in the **18th–20th centuries**, corresponding with modern artistic expansion and museum collecting practices.
- Several artists appear repeatedly in the collection, highlighting the museum's focus on influential figures in art history.

## Dataset

This project uses the **Metropolitan Museum of Art Open Access dataset**, which contains information on over **480,000 artworks** in the museum’s collection.

Key fields include:

- Department
- Country of origin
- Artistic medium
- Artist name
- Object begin date

The dataset allows exploration of historical, geographic, and material trends across the museum’s holdings.

## Artwork Production Across Centuries

To understand historical trends in artistic production, this analysis groups artworks by century using the object begin date.

```python
df["Object Begin Date"].describe()
```

```python
df_clean = df[(df["Object Begin Date"] > -5000) & (df["Object Begin Date"] < 2025)]
```

```python
df_clean.shape
```

```python
df_clean["Century"] = (df_clean["Object Begin Date"] // 100 + 1) * 100
```

```python
df_clean = df[(df["Object Begin Date"] > -5000) & (df["Object Begin Date"] < 2025)].copy()

df_clean["Century"] = (df_clean["Object Begin Date"] // 100 + 1) * 100
```

```python
century_counts = df_clean["Century"].value_counts().sort_index()

century_counts.head(20)
```

```python
import matplotlib.pyplot as plt

century_counts.plot(figsize=(12,6))

plt.title("Artworks in the Met Collection by Century")
plt.xlabel("Century")
plt.ylabel("Number of Artworks")
plt.show()
```

## Most Represented Artists in the Collection

This analysis identifies the artists with the largest number of works represented in the Metropolitan Museum of Art dataset.

```python
artists = df[df["Artist Display Name"].notna()]
artists = artists[artists["Artist Display Name"] != ""]
```

```python
top_artists = artists["Artist Display Name"].value_counts().head(10)

top_artists
```

```python
top_artists = artists[artists["Artist Display Name"] != "Unknown"]["Artist Display Name"].value_counts().head(10)

top_artists
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12,6))

ax.bar(top_artists.index, top_artists.values)

ax.set_title("Artists with the Most Works in the Met Collection")
ax.set_xlabel("Artist")
ax.set_ylabel("Number of Artworks")

plt.xticks(rotation=45, ha="right")

fig.tight_layout()

fig.savefig("images/top_artists_chart.png", dpi=300, bbox_inches="tight")

plt.show()
```
